```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
database=pd.read_csv("forestfires.csv")
database
```

```python
covarianza1=database['RH']
covarianza2=database['wind']
```

```python
covarianza2.cov(covarianza1)
```

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(covarianza2, covarianza1)
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, covarianza2))

plt.scatter(covarianza2, covarianza1)
plt.plot(covarianza2, mymodel)
plt.show()
```

# Resultados
Al examinar el valor de la covarianza se observa que no existe una relacion muy fuerte entre las dos variaves RH y wind. Esto ya que si se examinan las variables no existe una relacion inversa o directamente proporcional. Esto se debe a que varios valores de RH presenta en su contraparte varios valores en la variable RH; esto se presenta de la forma inversa; es decir para un valor de Rh hay valores diferentes
