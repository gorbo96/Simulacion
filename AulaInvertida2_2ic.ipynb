```python
from scipy.stats import pareto,t,chi2,beta,gamma,lognorm
import numpy as np 
import matplotlib.pyplot as plt
k = 2.3 # parametro de forma.
pareto = pareto(k)
x = np.linspace(pareto.ppf(0.01),
                pareto.ppf(0.99), 100)
fp = pareto.pdf(x) # Funcin de Probabilidad
plt.plot(x, fp)
plt.title('Distribucin de Pareto')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
# histograma
aleatorios = pareto.rvs(1000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma de Pareto')
plt.show()
```
