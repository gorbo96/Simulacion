```python
import pandas as pd
import numpy as np
import matplotlib.image as pltimg
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv("datos.csv",sep=";")
datos
```

```python
panamaS=datos.loc[datos["No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS (***)"]!=0].count()[0]
panamaN=datos.loc[datos["No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS (***)"]==0].count()[0]
etiquetas = ["Tiene","No tiene"]
valores = [panamaS,panamaN]
plt.figure()
barras = plt.bar(etiquetas, valores)
plt.show() 
```

```python
bajado=datos.loc[datos["Variación Ranking Tamaño 2020 - 2021"]<0].count()[0]
subido=datos.loc[datos["Variación Ranking Tamaño 2020 - 2021"]>0].count()[0]
etiquetas = ["Subido","Bajado"]
valores = [subido,bajado]
plt.figure()
barras = plt.bar(etiquetas, valores)
plt.show()  
```

```python
aporte=datos.loc[datos["Ranking (1) Tamaño  2021"]<10]
aporte
```

```python
import re
valores=[]
for porcentaje in aporte["Contribución   I.R. 2020"]:
  valores.append(float(re.sub("\%","",porcentaje)))
valores
```

```python
etiquetas=[]
for nombres in aporte["Nombre Grupo Económico"].values:
  etiquetas.append(nombres.split("(")[0])
etiquetas
```

```python
plt.pie(valores, labels=etiquetas, autopct='%.2f %%')
plt.show()
```
