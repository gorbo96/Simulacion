```python
import pandas as pd
import numpy as np
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import pareto
```

```python
k = 2.3 # parametro de forma.
paretol = pareto(k)
x = np.linspace(paretol.ppf(0.01),
                paretol.ppf(0.99), 100)
fp = paretol.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución de Pareto')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
aleatorios = paretol.rvs(100) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma de Pareto')
plt.show()
```

La cuantía por reclamación, X, (en miles de euros) en una compañía aseguradora
es una v.a. con distribución de Pareto de parámetros a= 3 (miles de euros) b=2,6.
La compañía decide reasegurarse , de manera que se cubra 6000 euros.

```python
aleatoriospython = []
[aleatoriospython.append(random.random())for i in range(100)]
salida=[]
for numero in aleatoriospython:
  salida.append(2.6/(numero**(1/3)))
x, y, z =plt.hist(salida)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma de Pareto')
plt.show()
```

```python
y=[]
for numero in salida:
  y.append(3*2.3**3/numero**4)
plt.scatter(salida, y)
plt.title('Distribución de Pareto')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
