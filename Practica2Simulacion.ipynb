```python
import pandas as pd
import numpy as np
import matplotlib.image as pltimg
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv("fifa_datos.csv",sep=",",encoding="utf-8")
datos.head()
```

```python
aux=datos['Weight'].fillna("000lbs")
rango1=0
rango2=0
rango3=0
rango4=0
for peso in aux:
  if(int(peso[0:3])>0 and int(peso[0:3])<125):
    rango1+=1
  if(int(peso[0:3])>=125 and int(peso[0:3])<150):
    rango2+=1
  if(int(peso[0:3])>=150 and int(peso[0:3])<175):
    rango3+=1
  if(int(peso[0:3])>=175):
    rango4+=1
plt.figure()
etiquetas = ['<125', '125-149','150-174','>174']
plt.pie([rango1, rango2,rango3,rango4], labels=etiquetas, autopct='%.2f %%')
plt.title('Porcetanje por rango de pesos')
plt.show()
```

```python
aux=datos['Overall']
rango1=0
rango2=0
rango3=0
rango4=0
rango5=0
rango6=0
for peso in aux:
  if(peso>=40 and peso<50):
    rango1+=1
  if(peso>=50 and peso<60):
    rango2+=1
  if(peso>=60 and peso<70):
    rango3+=1
  if(peso>=70 and peso<80):
    rango4+=1
  if(peso>=80 and peso<90):
    rango5+=1  
  if(peso>=90):
    rango6+=1
plt.figure()
etiquetas = ['40', '50','60','70','80','90']
plt.bar(etiquetas,[rango1, rango2,rango3,rango4,rango5,rango6])
plt.title('Conteo segun habilidad')
plt.show()
```

```python
import altair as alt

df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv')
df
```

```python
countries = ['United Kingdom', 'Germany', 'France']

subset = df[df.location.isin(countries)]
subset
```

```python
base = alt.Chart(subset).encode(
    x='date:T',
    color='location'
).mark_area(
    opacity=0.5, 
    stroke='black'
).interactive()
```

```python
c1 = base.encode(y='total_vaccinations_per_hundred')
c2 = base.encode(y='daily_vaccinations_per_million')
c3 = base.encode(y='people_vaccinated')
```

```python
import datapane as dp
report = dp.Report(
  dp.DataTable(df, caption=f'Dataset for {countries}'),
  dp.Group(dp.Plot(c1, caption='Vaccination %'), dp.Plot(c2, caption='Daily vaccinations per million'), columns=2),
  dp.Plot(c3, caption='Total vaccinations')
)
```
